```python
import requests
import re
from bs4 import BeautifulSoup
```

### ① 取出今天所有的發文

```python
url = 'https://www.ettoday.net/news/news-list.htm'
headers = {'user-agent': 'my-app/0.0.1'}
htmlfile = requests.get(url, headers=headers)

soup = BeautifulSoup(htmlfile.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('em')[-1].text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
# class dict
class_dict = {}
# {類別: [文章標題]}
for d in soup.find(class_="part_list_2").find_all('h3'):
    if d.find_all('em')[-1].text in class_dict.keys(): # 已經出現過的文章類別
        class_dict[d.find_all('em')[-1].text].append(d.find_all('a')[-1].text)
    else: # 第一次出現的文章類別
        class_dict[d.find_all('em')[-1].text] = [d.find_all('a')[-1].text]
               
class_dict
```

### ③ 哪一個類別的文章最多

```python
count = {key : len(value) for key,value in class_dict.items()}
count
```

```python
max(count.items())
```
