### 靜態爬蟲

```python
# 先複習一下原本純靜態的爬法

import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='./chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
# 每個兩秒鐘自動往下滑

import time
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
    
# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)

# 關閉瀏覽器
browser.quit();
```

## 2. 取出現在時間兩小時內的新聞

```python
import datetime

for d in soup.find(class_="part_list_2").find_all('h3'):
    date = datetime.datetime.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M")
    if date.hour >= 15 and date.hour <= 17:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
import time
today_year = time.strftime("%Y", time.localtime())
today_month = time.strftime("%m", time.localtime())
today_day = time.strftime("%d", time.localtime())
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import datetime

# 計算日期
before_day = 3
day_now = datetime.datetime.now()
delta = datetime.timedelta(days=before_day)
n_days = day_now - delta
today_year = n_days.strftime('%Y')
today_month = n_days.strftime('%m')
today_day = n_days.strftime('%d')
print(today_year, today_month, today_day)

url = 'https://www.ettoday.net/news/news-list-{}-{}-{}-0.htm'.format(today_year, today_month, today_day)
print(url)
browser = webdriver.Chrome(executable_path='./chromedriver')
browser.get(url)
# 每個兩秒鐘自動往下滑

import time
for i in range(18):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")

# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")


for d in soup.find(class_="part_list_2").find_all('h3'):
    date = datetime.datetime.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M")
    if date.hour >= 15 and date.hour <= 17:
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
 
# 關閉瀏覽器
browser.quit();
```
