### 1. Dcard ç¶²å€ï¼š https://www.dcard.tw/f

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.dcard.tw/f'

'''
Your Code Here
'''
headers = {'user-agent': 'my-app/0.0.1'}
htmlfile = requests.get(url, headers=headers)

htmlfile.encoding = 'utf-8'
print(htmlfile.text[0:3000])
```

```python
print('Request å–å›ä¹‹å¾Œè©²æ€éº¼å–å‡ºè³‡æ–™ï¼Œè³‡æ–™å‹æ…‹æ˜¯ä»€éº¼ï¼Ÿ =>')
type(htmlfile.text)
```

```python
print('ç‚ºä»€éº¼è¦ä½¿ç”¨ BeatifulSoup è™•ç†ï¼Ÿè™•ç†å¾Œçš„å‹æ…‹æ˜¯ä»€éº¼ï¼Ÿ => ')
soup = BeautifulSoup(htmlfile.text, "lxml")
print(type(soup))
```

```python
posts = soup.find_all('a', class_='PostEntry_root_V6g0rd')
for post in posts:
    post_title = post.find('h3', class_='Title__Text-v196i6-0 gmfDU')
    if post_title:
        print(post_title.text)
```

```python
soup = BeautifulSoup(r.text, "html5lib")
# print(soup.prettify())

posts = soup.find_all('a', class_='ExploreSpecialCard-title')
for post in posts:
    print(post.text)
```
